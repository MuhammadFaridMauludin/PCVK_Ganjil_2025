```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('ktm.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# cara Split dan Merge channel RGB
R, G, B = cv2.split(img)

rb = cv2.merge([R*1, G*0, B*1])
gb = cv2.merge([R*0, G*1, B*1])

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(rb)
plt.title("Channel Red Blue")

plt.subplot(1,2,2)
plt.imshow(gb)
plt.title("Channel Green Blue")
plt.show()
```

```python
img = cv2.imread('ktm.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

R, G, B = cv2.split(img)

red = cv2.merge([R*1, G*0, B*1])
green = cv2.merge([R*0, G*1, B*1])
blue =  cv2.merge([R*0, G*0, B*1])
orange = cv2.merge([R*1, G*1, B*1])
yellow = cv2.merge([R*1, G*1, B*0])

red_crop = red[5:100, 25:140]
green_crop = green[10:150, 50:140]
blue_crop = blue[20:150, 100:140]
orange_crop = orange[50:120, 50:60]
yellow_crop = yellow[500:1800, 800:1200]

plt.figure(figsize=(12,5))

plt.subplot(1,5,1)
plt.imshow(red_crop)
plt.title("potongan red")

plt.subplot(1,5,2)
plt.imshow(green_crop)
plt.title("potongan green")

plt.subplot(1,5,3)
plt.imshow(blue_crop)
plt.title("potongan blue")

plt.subplot(1,5,4)
plt.imshow(orange_crop)
plt.title("potongan orange")

plt.subplot(1,5,5)
plt.imshow(yellow_crop)
plt.title("potongan yellow")

plt.show()
```

```python
img = cv2.imread('ktm.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

R, G, B = cv2.split(img)

red = cv2.merge([R*1, G*0, B*1])
green = cv2.merge([R*0, G*1, B*1])
blue =  cv2.merge([R*0, G*0, B*1])
orange = cv2.merge([R*1, G*1, B*1])
yellow = cv2.merge([R*1, G*1, B*0])

red_crop = red[5:100, 25:140]
green_crop = green[10:150, 50:140]
blue_crop = blue[20:150, 100:140]
orange_crop = orange[50:120, 50:60]
yellow_crop = yellow[500:1800, 500:700]

plt.figure(figsize=(12,5))

redBalik=cv2.flip(red_crop,1)
greenBalik=cv2.flip(green_crop,1)
blueBalik=cv2.flip(blue_crop,1)
orangeBalik=cv2.flip(orange_crop,1)
yellowBalik=cv2.flip(yellow_crop,1)

plt.subplot(1,5,1)
plt.imshow(redBalik)
plt.title("potongan red  flip 1")

plt.subplot(1,5,2)
plt.imshow(greenBalik)
plt.title("potongan green  flip 1")

plt.subplot(1,5,3)
plt.imshow(blueBalik)
plt.title("potongan blue flip 1")

plt.subplot(1,5,4)
plt.imshow(orangeBalik)
plt.title("potongan orange flip 1")

plt.subplot(1,5,5)
plt.imshow(yellowBalik)
plt.title("potongan yellow flip 1")

plt.show()
```

```python
img = cv2.imread('myfoto.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

R, G, B = cv2.split(img)
bl = cv2.merge([R*0, G*0, B*1])

plt.figure(figsize=(20,10))

cv2.rectangle(bl,pt1=(400,460),pt2=(310,340), color=(0,255,0),thickness=5)
cv2.circle(bl,center=(353,388),radius=3,color=(255,0,0),thickness=5)
cv2.circle(bl,center=(358,378),radius=3,color=(255,255,255),thickness=5)
cv2.circle(bl,center=(347,378),radius=3,color=(255,255,255),thickness=5)
cv2.rectangle(bl,pt1=(343,398),pt2=(365,400), color=(255,255,255),thickness=5)

cv2.rectangle(bl,pt1=(250,895),pt2=(5,795), color=(255,255,255),thickness=5)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(bl,text='myfoto',org=(15,865),fontFace=font,fontScale=2,color=(255,255,0),thickness=3, lineType=cv2.LINE_AA)

#plt.grid()
plt.imshow(bl)
plt.show()
```
