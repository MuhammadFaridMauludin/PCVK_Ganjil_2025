```python
#menampilkan gambar dengan plotting
import cv2
import matplotlib.pyplot as plt #untuk plotting
from google.colab.patches import cv2_imshow
img = cv2.imread('ktm.jpg')#nama file sesuai dengan yang diaplud
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #mengubah warna gbr menjadi rgb
plt.imshow(img)#menampilkan gambar
```

```python
#menampilkan citra grayscale
img_gray = cv2.imread('ktm.jpg', cv2.IMREAD_GRAYSCALE)

#tanpa cmap
plt.subplot(1,3,1) # untuk mengatur grid misal baris 1 kolom 2 gambar 1
plt.imshow(img_gray)#membaca gambar sebelum di tampilkan
plt.title('tanpa Cmap')

#dengan cmap
plt.subplot(1,3,2)
plt.imshow(img_gray, cmap='gray')
plt.title('dengan Cmap')

#ubah warna cmap
plt.subplot(1,3,3)
plt.imshow(img_gray, cmap='magma')
plt.title('dengan Cmap')

plt.show()
```

```python
# melakukan resizing
img = cv2.imread('ktm.jpg')#nama file sesuai dengan yang diaplud
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #mengubah warna gbr menjadi rgb
imgRGBsmall=cv2.resize(img,(400,250))#mengubah ukuran pixel menjadi lebar 400 tinggi 250
plt.imshow(imgRGBsmall)
plt.show()
```

```python
#melakukan flipping
imgBalik=cv2.flip(imgRGBsmall,1)
plt.imshow(imgBalik)
plt.show()
#parameter flip
#0 flip vertikal
#1 flip horizontal
#-1 flip keduanya
```

```python
#untuk memperbesar ukuran canvas dengan posisi gambar terbaik
imgBalik=cv2.flip(imgRGBsmall,1)

#tampilkan ukuran canvas yang lebih besar
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111)
ax.imshow(imgBalik)
plt.show()
```

```python
#membuat bentuk 2d dari open cv
import cv2
import numpy as np#import library numpy
import matplotlib.pyplot as plt#import plotting

#np.zeros membuat array dengan elemen bernilai 0
#shape ukuran gambar(3 itu jumlah warna)
#dtype menentukan type data elemen untuk menampung warna
black_img = np.zeros(shape=(510,510,3),dtype=np.int16)
plt.imshow(black_img)
plt.show()
```

```python
#perhatikan koordinat titip 2 pt1 dan pt2
cv2.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
plt.show()
```

```python
#perhatikan koordinat titip 2 pt1 dan pt2
cv2.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
plt.show()
```

```python
cv2.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
plt.show()
```

```python
cv2.line(black_img,pt1=(0,0),pt2=(510,510),color=(255,255,255),thickness=5)
plt.imshow(black_img)
plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img,text='hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2, lineType=cv2.LINE_AA)
plt.imshow(black_img)
plt.show()
```
