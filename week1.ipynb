```python
# library yang digunakan
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as pit
```

```python
# membuat list untuk menyimpan url dari beberapa image
urls = [
"https://radartv.disway.id/upload/58585f7adeefcceb3316172e315b9c6e.jpg",
"https://static.promediateknologi.id/crop/0x0:0x0/750x500/webp/photo/p1/408/2025/02/18/C1C24799-D2EF-48F2-83A7-C0BE64E41DD3-3901725431.png"]
#baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
#read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
#resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
#convert color to RGB
  final_frame = cv.hconcat((image, image_2))
#concatenate image
  cv2_imshow(final_frame)
#show image]
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]

print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)

cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image
for x in range(tinggi):
    image_3[x, int((lebar)/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal dari kiri-atas ke kanan-bawah
for x in range(lebar):
    y = int((tinggi / lebar) * x)
    if y < tinggi:
        image_3[y, x] = [255, 255, 255]

# (opsional) Garis dari kanan-atas ke kiri-bawah
for x in range(lebar):
    y = int(tinggi - ((tinggi / lebar) * x) - 1)
    if y >= 0:
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal tapi hanya sebagian di tengah
start = int(lebar * 0.50)
end   = int(lebar * 0.70)
row   = int(tinggi / 2)

for y in range(start, end):
    image_3[row, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
#ukuran kotak
ulebar  = int(lebar * 0.3)
utinggi = int(tinggi * 0.2)
# posisi kotak
tengah_x = lebar // 2
tengah_y = tinggi // 2
#batas kotak
x1 = tengah_x - ulebar // 2
x2 = tengah_x + ulebar // 2
y1 = tengah_y - utinggi // 2
y2 = tengah_y + utinggi // 2
image_3[y1:y2, x1:x2] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
